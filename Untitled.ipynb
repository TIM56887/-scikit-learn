```python
import pandas as pd
import numpy as np
import joblib
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

```

```python
#欄位名稱：district(鄉鎮市區)、rps01(交易標的)、rps02(土地區段位置建物區段門牌)、rps03(土地移轉總面積平方公尺)、rps04(都市土地使用分區)、
#rps05(非都市土地使用分區)、rps06(非都市土地使用編定)、rps07(交易年月日)、rps08(交易筆棟數)、rps09(移轉層次)、rps10(總樓層數)、
#rps11(建物型態)、rps12(主要用途)、rps13(主要建材)、rps14(建築完成年月)、rps15(建物移轉總面積平方公尺)、rps16(建物現況格局-房)、
#rps17(建物現況格局-廳)、rps18(建物現況格局-衛)、rps19(建物現況格局-隔間)、rps20(有無管理組織)、rps21(總價元)、rps22(單價元平方公尺)、
#rps23(車位類別)、rps24(車位移轉總面積平方公尺)、rps25(車位總價元)、rps26(備註)、rps27(編號)、rps28(主建物面積)、rps29(附屬建物面積)、
#rps30(陽台面積)、rps31(電梯)
df = pd.read_csv("data.csv")
df
```

```python
#排除只有土地或車位和不是住宅用的交易
df = df[df['rps01'].isin(['房地(土地+建物)', '房地(土地+建物)+車位','建物'])]
df = df[df['rps04'].isin(['住'])]
df = df[df['rps12'].isin(['住家用','住商用','住辦用'])]
df
```

```python
# 篩選要用於訓練的特徵
df = df[['district', 'rps03','rps09','rps10','rps11','rps14','rps15','rps16','rps17','rps18','rps19','rps20','rps31','rps21']]
df
```

```python
# 將rps09、rps10欄位的樓層轉換為數字
def chinese_to_arabic(cn_num):
    try:
        digits = {'零': 0, '一': 1, '二': 2, '三': 3, '四': 4, '五': 5, '六': 6, '七': 7, '八': 8, '九': 9, '十': 10}

        # 檢查輸入是否為 NaN 或非字符串
        if isinstance(cn_num, float) and np.isnan(cn_num):
            return cn_num
        if not isinstance(cn_num, str):
            return None

        if len(cn_num) == 3 and cn_num[1] == '十':
            ten_digit = digits[cn_num[0]] * 10
            unit_digit = digits[cn_num[2]]
            return ten_digit + unit_digit
        elif len(cn_num) == 2 and cn_num[1] == '十':
            return digits[cn_num[0]] * 10
        elif len(cn_num) == 2:
            return 10 + digits[cn_num[1]]
        elif cn_num == '十':
            return 10
        else:
            return digits[cn_num]
    except:
        return cn_num
df['rps09'] = df['rps09'].apply(lambda x: chinese_to_arabic(str(x)[:-1]) if isinstance(x, str) else float(x))
df['rps10'] = df['rps10'].apply(lambda x: chinese_to_arabic(str(x)[:-1]) if isinstance(x, str) else float(x))
df['rps10'] = df['rps10'].astype(int, errors='ignore')

df
```

```python
# 計算屋齡
def houseold(years):
    str_years = str(years)
    if '.' in str_years:
        str_years = str(int(float(str_years)))  # Convert float string to int string
    if len(str_years) == 6:
        return 111 - int(str_years[:2])
    elif len(str_years) ==7:
        return 111 - int(str_years[:3])
    else:
        return years

df['rps14'] = df['rps14'].apply(houseold)
df
```

```python
# 刪掉統計不完整的資料
df['rps10'] = pd.to_numeric(df['rps10'], errors='coerce')
df = df.dropna(subset=['rps10'])
df['rps09'] = pd.to_numeric(df['rps09'], errors='coerce')
df = df.dropna(subset=['rps09'])
df = df.dropna(subset=['rps14'])
```

```python
# 拆分訓練資料
x = df[['district', 'rps03','rps09','rps10','rps11','rps14','rps15','rps16','rps17','rps18','rps19','rps20','rps31']].values
y = df[['rps21']].values
```

```python
# 將縣市使用one hot encoding，其他使用label encoding
enc = OneHotEncoder(sparse=False)
onehotlabels = enc.fit_transform(x[:,0].reshape(-1, 1))



x = np.delete(x, 0, 1)


x = np.concatenate((onehotlabels, x), axis=1)
```

```python
typeencoder =LabelEncoder()
Le = LabelEncoder()
le_list = [34,35,36]
```

```python

for i in le_list:
    x[:,i] = Le.fit_transform(x[:, i])

x[:,28] = typeencoder.fit_transform(x[:,28])
```

```python
lreg = linear_model.LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

```python
model = lreg.fit(X_train, y_train)
predictions = model.predict(X_test)
print("predictions: ",predictions)







print("R^2 value: ", lreg.score(x,y))
print("權重: ",lreg.coef_) 
print("intercept: ",lreg.intercept_)
```
